```python
# ============================================================
# STEP 1 — LOAD & INSPECT THE DATA
# ============================================================
 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
 
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
```

```python
# --- Load ---
df = pd.read_csv('coca_cola.csv', parse_dates=['Date'])
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date').reset_index(drop=True)
 
print("Shape:", df.shape)
print("\nColumns:\n", df.columns.tolist())
print("\nData types:\n", df.dtypes)
print("\nFirst 5 rows:\n", df.head())
print("\nBasic stats:\n", df.describe().round(2))
```

```python
# --- Check missing values ---
missing = df.isnull().sum()
print("\nMissing values:\n", missing[missing > 0])
# NOTE: MA_200 will have ~199 NaN rows at the start — this is normal
```

```python
# ============================================================
# STEP 2 — CLEAN THE DATA
# ============================================================

df.dropna(subset=['MA_200', 'RSI_14', 'BB_Upper'], inplace=True)
df.reset_index(drop=True, inplace=True)

# ── Fix Day_of_Week ──
day_map = {'Monday':0, 'Tuesday':1, 'Wednesday':2, 'Thursday':3,
           'Friday':4, 'Saturday':5, 'Sunday':6}
df['Day_of_Week'] = df['Day_of_Week'].map(day_map)

print(f"\nRows after dropping NaN: {len(df)}")

print(f"Date range: {df['Date'].min().date()} → {df['Date'].max().date()}")
```

```python
# ============================================================
# STEP 3 — EXPLORATORY DATA ANALYSIS (EDA)
# ============================================================
 
fig, axes = plt.subplots(4, 1, figsize=(14, 16), sharex=True)
fig.suptitle('Stock Price Overview', fontsize=16, fontweight='bold')
 
# Plot 1: Price + Moving Averages
axes[0].plot(df['Date'], df['Close'],  label='Close',  linewidth=1.2)
axes[0].plot(df['Date'], df['MA_20'],  label='MA 20',  linewidth=1, linestyle='--')
axes[0].plot(df['Date'], df['MA_50'],  label='MA 50',  linewidth=1, linestyle='--')
axes[0].plot(df['Date'], df['MA_200'], label='MA 200', linewidth=1, linestyle='--')
axes[0].fill_between(df['Date'], df['BB_Upper'], df['BB_Lower'], alpha=0.1, label='Bollinger Bands')
axes[0].set_ylabel('Price')
axes[0].legend(loc='upper left', fontsize=8)


# Plot 2: Volume
axes[1].bar(df['Date'], df['Volume'], width=1, alpha=0.6)
axes[1].plot(df['Date'], df['Volume_MA_20'], color='orange', label='Volume MA20', linewidth=1)
axes[1].set_ylabel('Volume')
axes[1].legend(fontsize=8)


# Plot 3: RSI
axes[2].plot(df['Date'], df['RSI_14'], color='purple', linewidth=1)
axes[2].axhline(70, color='red',   linestyle='--', linewidth=0.8, label='Overbought (70)')
axes[2].axhline(30, color='green', linestyle='--', linewidth=0.8, label='Oversold (30)')
axes[2].set_ylabel('RSI (14)')
axes[2].legend(fontsize=8)


# Plot 4: MACD
axes[3].plot(df['Date'], df['MACD'],        label='MACD',   linewidth=1)
axes[3].plot(df['Date'], df['MACD_Signal'], label='Signal', linewidth=1, linestyle='--')
macd_hist = df['MACD'] - df['MACD_Signal']
axes[3].bar(df['Date'], macd_hist, width=1, alpha=0.4,
            color=np.where(macd_hist >= 0, 'green', 'red'), label='Histogram')
axes[3].set_ylabel('MACD')
axes[3].legend(fontsize=8)
 
plt.tight_layout()
plt.savefig('step3_price_overview.png', dpi=120, bbox_inches='tight')
plt.show()
```

```python
# --- Daily return distribution ---
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
 
axes[0].hist(df['Daily_Return_Pct'], bins=80, edgecolor='none', color='steelblue', alpha=0.8)
axes[0].axvline(0, color='red', linestyle='--')
axes[0].set_title('Distribution of Daily Returns (%)')
axes[0].set_xlabel('Return (%)')
axes[0].set_ylabel('Count')


# Q-Q plot to check fat tails
from scipy import stats
stats.probplot(df['Daily_Return_Pct'].dropna(), dist="norm", plot=axes[1])
axes[1].set_title('Q-Q Plot (fat tails = heavy tails in returns)')
 
plt.tight_layout()
plt.savefig('step3_return_distribution.png', dpi=120, bbox_inches='tight')
plt.show()
```

```python
# --- Seasonality ---
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
 
monthly = df.groupby('Month')['Daily_Return_Pct'].mean()
axes[0].bar(monthly.index, monthly.values, color='teal', alpha=0.8)
axes[0].set_title('Average Return by Month')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Avg Daily Return (%)')
axes[0].axhline(0, color='red', linestyle='--')
 
dow_map = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri'}
daily = df.groupby('Day_of_Week')['Daily_Return_Pct'].mean()
axes[1].bar([dow_map.get(i, i) for i in daily.index], daily.values, color='coral', alpha=0.8)
axes[1].set_title('Average Return by Day of Week')
axes[1].set_xlabel('Day')
axes[1].set_ylabel('Avg Daily Return (%)')
axes[1].axhline(0, color='red', linestyle='--')
 
plt.tight_layout()
plt.savefig('step3_seasonality.png', dpi=120, bbox_inches='tight')
plt.show()
```

```python
# ============================================================
# STEP 4 — CORRELATION ANALYSIS
# ============================================================
 
# Select numeric columns only (exclude Date and the target we'll create later)
numeric_cols = ['Close', 'Volume', 'Daily_Return_Pct', 'Daily_Range',
                'MA_20', 'MA_50', 'MA_200', 'EMA_12', 'EMA_26',
                'MACD', 'MACD_Signal', 'BB_Upper', 'BB_Lower', 'BB_Width',
                'RSI_14', 'Volume_MA_20', 'Cumulative_Return_Pct']
 
corr = df[numeric_cols].corr()
 
plt.figure(figsize=(14, 10))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='RdYlGn',
            vmin=-1, vmax=1, center=0, linewidths=0.3,
            annot_kws={'size': 7})
plt.title('Correlation Matrix — Stock Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('step4_correlation.png', dpi=120, bbox_inches='tight')
plt.show()
 
# TIP: MA_20, MA_50, MA_200, EMA_12, EMA_26 will be highly correlated with Close.
# Use ratios instead of raw values to reduce multicollinearity.
```

```python
# ============================================================
# STEP 5 — FEATURE ENGINEERING
# ============================================================
 
def engineer_features(df):
    d = df.copy()
 
    # ── Ratio features (reduce multicollinearity) ──
    d['close_to_ma20']  = d['Close'] / d['MA_20']  - 1   # % deviation from MA20
    d['close_to_ma50']  = d['Close'] / d['MA_50']  - 1
    d['close_to_ma200'] = d['Close'] / d['MA_200'] - 1
    d['ma20_to_ma50']   = d['MA_20'] / d['MA_50']  - 1   # Golden/Death cross signal
    d['ma50_to_ma200']  = d['MA_50'] / d['MA_200'] - 1
 
    # ── MACD histogram (more informative than raw MACD) ──
    d['macd_hist']      = d['MACD'] - d['MACD_Signal']
    d['macd_cross']     = np.sign(d['macd_hist'])         # +1 bullish, -1 bearish
 
    # ── Bollinger %B: where is price within the bands? ──
    d['pct_b'] = (d['Close'] - d['BB_Lower']) / (d['BB_Upper'] - d['BB_Lower'])
    # 0 = at lower band, 0.5 = at midline, 1 = at upper band
 
    # ── Volume ratio ──
    d['volume_ratio'] = d['Volume'] / d['Volume_MA_20']   # >1 = high volume day
 
    # ── RSI zones (binary flags) ──
    d['rsi_overbought'] = (d['RSI_14'] > 70).astype(int)
    d['rsi_oversold']   = (d['RSI_14'] < 30).astype(int)
 
    # ── Trend regime ──
    d['above_ma200']    = (d['Close'] > d['MA_200']).astype(int)
    d['golden_cross']   = (d['MA_50'] > d['MA_200']).astype(int)
 
    # ── Lag features (give model "memory") ──
    for col in ['Daily_Return_Pct', 'RSI_14', 'macd_hist', 'volume_ratio', 'Close']:
        d[f'{col}_lag1'] = d[col].shift(1)
        d[f'{col}_lag2'] = d[col].shift(2)
        d[f'{col}_lag3'] = d[col].shift(3)
 
    # ── Rolling stats ──
    d['return_rolling_mean5']  = d['Daily_Return_Pct'].shift(1).rolling(5).mean()
    d['return_rolling_std5']   = d['Daily_Return_Pct'].shift(1).rolling(5).std()
    d['return_rolling_mean10'] = d['Daily_Return_Pct'].shift(1).rolling(10).mean()
 
    return d
 
df = engineer_features(df)
print(f"\nShape after feature engineering: {df.shape}")
 
```

```python
# ============================================================
# STEP 6 — CREATE TARGET VARIABLE
# ============================================================
 
# Target: Will tomorrow's closing price be HIGHER than today's?
# 1 = price goes up, 0 = price goes down or stays flat
df['target'] = (df['Close'].shift(-1) > df['Close']).astype(int)
 
# Drop the last row (no target available for it)
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
 
print(f"\nTarget distribution:")
print(df['target'].value_counts())
print(f"Up days: {df['target'].mean()*100:.1f}%  |  Down days: {(1-df['target'].mean())*100:.1f}%")
```

```python
# ============================================================
# STEP 7 — PREPARE FEATURES FOR ML
# ============================================================
 
FEATURES = [
    # Ratio-based trend features
    'close_to_ma20', 'close_to_ma50', 'close_to_ma200',
    'ma20_to_ma50', 'ma50_to_ma200',
 
    # Momentum
    'RSI_14', 'macd_hist', 'macd_cross',
    'rsi_overbought', 'rsi_oversold',
 
    # Volatility
    'BB_Width', 'pct_b',
 
    # Volume
    'volume_ratio',
 
    # Regime flags
    'above_ma200', 'golden_cross',
 
    # Lags (memory)
    'Daily_Return_Pct_lag1', 'Daily_Return_Pct_lag2', 'Daily_Return_Pct_lag3',
    'RSI_14_lag1', 'macd_hist_lag1', 'volume_ratio_lag1',
 
    # Rolling stats
    'return_rolling_mean5', 'return_rolling_std5', 'return_rolling_mean10',
 
    # Calendar
    'Month', 'Day_of_Week', 'Quarter',
]
 
X = df[FEATURES]
y = df['target']
 
print(f"\nFinal feature matrix: {X.shape}")
print(f"Missing values in X: {X.isnull().sum().sum()}")
```

```python
# ============================================================
# STEP 8 — TRAIN / TEST SPLIT (CHRONOLOGICAL — NEVER RANDOM!)
# ============================================================
 
split_idx = int(len(df) * 0.80)   # 80% train, 20% test
 
X_train = X.iloc[:split_idx]
X_test  = X.iloc[split_idx:]
y_train = y.iloc[:split_idx]
y_test  = y.iloc[split_idx:]
 
print(f"\nTrain: {len(X_train)} rows  ({df['Date'].iloc[0].date()} → {df['Date'].iloc[split_idx-1].date()})")
print(f"Test:  {len(X_test)}  rows  ({df['Date'].iloc[split_idx].date()} → {df['Date'].iloc[-1].date()})")
```

```python
# ============================================================
# STEP 9 — TRAIN MODELS
# ============================================================
 
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import (classification_report, confusion_matrix,
                              roc_auc_score, accuracy_score)
import xgboost as xgb
 
results = {}
 
# ── Model A: Logistic Regression (baseline) ──────────────────
lr_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('lr', LogisticRegression(C=0.1, max_iter=1000, random_state=42))
])
lr_pipe.fit(X_train, y_train)
lr_preds = lr_pipe.predict(X_test)
lr_probs = lr_pipe.predict_proba(X_test)[:, 1]
results['Logistic Regression'] = {
    'preds': lr_preds, 'probs': lr_probs,
    'acc': accuracy_score(y_test, lr_preds),
    'auc': roc_auc_score(y_test, lr_probs)
}

# ── Model B: Random Forest ────────────────────────────────────
rf = RandomForestClassifier(
    n_estimators=300, max_depth=6,
    min_samples_leaf=20, random_state=42, n_jobs=-1
)
rf.fit(X_train, y_train)
rf_preds = rf.predict(X_test)
rf_probs = rf.predict_proba(X_test)[:, 1]
results['Random Forest'] = {
    'preds': rf_preds, 'probs': rf_probs,
    'acc': accuracy_score(y_test, rf_preds),
    'auc': roc_auc_score(y_test, rf_probs)
}
 
# ── Model C: XGBoost ─────────────────────────────────────────
xgb_model = xgb.XGBClassifier(
    n_estimators=500, max_depth=4, learning_rate=0.03,
    subsample=0.8, colsample_bytree=0.8, min_child_weight=10,
    eval_metric='logloss', early_stopping_rounds=30,
    random_state=42, verbosity=0
)
xgb_model.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],
    verbose=False
)
xgb_preds = xgb_model.predict(X_test)
xgb_probs = xgb_model.predict_proba(X_test)[:, 1]
results['XGBoost'] = {
    'preds': xgb_preds, 'probs': xgb_probs,
    'acc': accuracy_score(y_test, xgb_preds),
    'auc': roc_auc_score(y_test, xgb_probs)
}
 
# ── Summary Table ──────────────────────────────────────────────
print("\n" + "="*50)
print("MODEL COMPARISON")
print("="*50)
for name, res in results.items():
    print(f"\n{name}")
    print(f"  Accuracy : {res['acc']*100:.2f}%")
    print(f"  ROC-AUC  : {res['auc']:.4f}")
    print(classification_report(y_test, res['preds'],
                                 target_names=['Down', 'Up'], digits=3))
 
```

```python
# ============================================================
# STEP 10 — EVALUATE & VISUALISE RESULTS
# ============================================================
 
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Model Evaluation', fontsize=14, fontweight='bold')
 
model_names = list(results.keys())
 
for i, (name, res) in enumerate(results.items()):
    cm = confusion_matrix(y_test, res['preds'])
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[i],
                xticklabels=['Down', 'Up'], yticklabels=['Down', 'Up'])
    axes[i].set_title(f"{name}\nAcc={res['acc']*100:.1f}%  AUC={res['auc']:.3f}")
    axes[i].set_xlabel('Predicted')
    axes[i].set_ylabel('Actual')
 
plt.tight_layout()
plt.savefig('step10_confusion_matrices.png', dpi=120, bbox_inches='tight')
plt.show()
 
# ROC curves
from sklearn.metrics import roc_curve
 
plt.figure(figsize=(7, 5))
for name, res in results.items():
    fpr, tpr, _ = roc_curve(y_test, res['probs'])
    plt.plot(fpr, tpr, label=f"{name} (AUC={res['auc']:.3f})", linewidth=2)
plt.plot([0,1],[0,1],'k--', linewidth=0.8, label='Random (0.500)')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves')
plt.legend()
plt.tight_layout()
plt.savefig('step10_roc_curves.png', dpi=120, bbox_inches='tight')
plt.show()
 
```

```python
# ============================================================
# STEP 11 — FEATURE IMPORTANCE (Random Forest & XGBoost)
# ============================================================
 
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
 
# Random Forest importance
rf_imp = pd.Series(rf.feature_importances_, index=FEATURES).sort_values(ascending=True)
rf_imp.tail(15).plot(kind='barh', ax=axes[0], color='steelblue', alpha=0.8)
axes[0].set_title('Random Forest — Top 15 Features')
axes[0].set_xlabel('Importance')

# XGBoost importance
xgb_imp = pd.Series(xgb_model.feature_importances_, index=FEATURES).sort_values(ascending=True)
xgb_imp.tail(15).plot(kind='barh', ax=axes[1], color='coral', alpha=0.8)
axes[1].set_title('XGBoost — Top 15 Features')
axes[1].set_xlabel('Importance')
 
plt.tight_layout()
plt.savefig('step11_feature_importance.png', dpi=120, bbox_inches='tight')
plt.show()
```

```python
# ============================================================
# STEP 12 — SIMPLE BACKTEST (Signal-based return simulation)
# ============================================================
 
# Use XGBoost predictions on the test set
test_df = df.iloc[split_idx:].copy()
test_df['signal']          = xgb_preds           # 1 = go long, 0 = stay out
test_df['strategy_return'] = test_df['signal'] * test_df['Daily_Return_Pct']
test_df['cum_strategy']    = (1 + test_df['strategy_return'] / 100).cumprod()
test_df['cum_buyhold']     = (1 + test_df['Daily_Return_Pct'] / 100).cumprod()
 
plt.figure(figsize=(12, 5))
plt.plot(test_df['Date'], test_df['cum_buyhold'],  label='Buy & Hold', linewidth=2)
plt.plot(test_df['Date'], test_df['cum_strategy'], label='ML Strategy (XGBoost)', linewidth=2)
plt.title('Backtest: ML Strategy vs Buy & Hold')
plt.xlabel('Date')
plt.ylabel('Cumulative Return (1 = starting capital)')
plt.legend()
plt.axhline(1, color='gray', linestyle='--', linewidth=0.8)
plt.tight_layout()
plt.savefig('step12_backtest.png', dpi=120, bbox_inches='tight')
plt.show()
```

```python
# Performance metrics
n_days      = len(test_df)
n_years     = n_days / 252
strat_ret   = test_df['strategy_return']
 
total_return = test_df['cum_strategy'].iloc[-1] - 1
annual_ret   = (1 + total_return) ** (1 / n_years) - 1
sharpe       = (strat_ret.mean() / strat_ret.std()) * np.sqrt(252) if strat_ret.std() != 0 else 0
win_rate     = (test_df[test_df['signal'] == 1]['strategy_return'] > 0).mean()
 
print("\n" + "="*40)
print("BACKTEST PERFORMANCE SUMMARY")
print("="*40)
print(f"Total return   : {total_return*100:.2f}%")
print(f"Annualised ret : {annual_ret*100:.2f}%")
print(f"Sharpe ratio   : {sharpe:.3f}")
print(f"Win rate       : {win_rate*100:.1f}%")
print(f"Signals fired  : {test_df['signal'].sum()} / {n_days} days")
print("="*40)
print("\nNOTE: This backtest ignores transaction costs, slippage,")
print("and market impact. Real-world returns will be lower.")
```

```python
# ============================================================
# STEP 13 — CROSS-VALIDATION (Walk-forward)
# ============================================================
 
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import roc_auc_score
 
tscv   = TimeSeriesSplit(n_splits=5)
scores = []
 
print("\nWalk-forward cross-validation (XGBoost):")
print("-" * 40)
 
for fold, (train_idx, val_idx) in enumerate(tscv.split(X)):
    X_tr, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_tr, y_val = y.iloc[train_idx], y.iloc[val_idx]
 
    m = xgb.XGBClassifier(
        n_estimators=200, max_depth=4, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8,
        eval_metric='logloss', verbosity=0, random_state=42
    )
    m.fit(X_tr, y_tr)
    proba = m.predict_proba(X_val)[:, 1]
    auc   = roc_auc_score(y_val, proba)
    acc   = accuracy_score(y_val, m.predict(X_val))
    scores.append(auc)
    print(f"Fold {fold+1}: AUC={auc:.4f}  Acc={acc*100:.1f}%")
 
print("-" * 40)
print(f"Mean AUC: {np.mean(scores):.4f} ± {np.std(scores):.4f}")
```
